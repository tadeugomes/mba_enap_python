#### Entrada 2: 6

```

  ooooooo
 ooooooooo
ooooooooooo
```


````python
prompt210="""def gerar_piramide(n):
    resultado = ""
    for i in range(1, n + 1):
        espacos = " " * (n - i)  # Adiciona espaços para centralizar
        caracteres = "o" * (2 * i - 1)  # Calcula o número de 'o'
        resultado += espacos + caracteres + "\n"
    return resultado.strip()

# Exemplos de uso
entrada1 = 5
print(gerar_piramide(entrada1))

entrada2 = 6
print(gerar_piramide(entrada2))"""
````

### 2.11 Crie um prompt de uma função python que retorne uma string com o seguinte padrão.

#### Entrada 1: 5

#### Resulatdo esperado 1:
```
ooooo
o   o
o   o
o   o
ooooo
```

#### Entrada 2: 6

#### Resulatdo esperado 2:
```
oooooo
o    o
o    o
o    o
oooooo
```

````python
prompt211="""def gerar_quadrado_vazado(n):
    resultado = ""
    for i in range(1, n + 1):
        if i == 1 or i == n:
            # Primeira e última linha: só "o"
            resultado += "o" * n + "\n"
        else:
            # Linhas intermediárias: "o" com espaços no meio
            resultado += "o" + " " * (n - 2) + "o" + "\n"
    return resultado.strip()

# Exemplos de uso
entrada1 = 5
print(gerar_quadrado_vazado(entrada1))

entrada2 = 6
print(gerar_quadrado_vazado(entrada2))"""
````

````python
validate(prompt211, "A2-E11")
````
